```python
data_route_copy = data_route.copy()
data_route_copy['Num_tracking'] = 1 
data_route_ing['date'] = pd.to_datetime(data_route_ing['date'])
data_routeForMap = data_route_ing.groupby(['date','province','city','latitude','longitude']).sum()
data_routeForMap = data_routeForMap.reset_index()
data_routeForMap
```

```python
data_GPS = data_route_copy[['id','latitude','longitude']] 
data_GPS
#data_routeForMap = pd.merge(['id',
```

```python
# Plus the contact_number of confirmed personregion to that region 
data_routeForMap = pd.merge(data_routeForMap,data_GPS, on=['latitude','longitude'])
data_routeForMap
```

```python
plus_info = data_patient[['id','contact_number']] 
plus_info = plus_info.fillna(0)
plus_info.rename(columns ={"id":"id_x"},inplace =True)
plus_info['id_y'] = plus_info['id_x']
plus_info
data_map = pd.merge(data_routeForMap,plus_info, on=['id_x','id_y'])
#data_routeForMap
```

```python
data_map
```

```python
data_map= data_map.drop_duplicates(['latitude','longitude','id_x','id_y'],keep='first')
data_map
```

```python
# World wide Map Plotting

m = folium.Map(location=[36.3004049, 127.7613334], tiles='cartodbpositron',
               min_zoom=5, max_zoom=8, zoom_start=40) # 줌인아웃 크기 조절 가능 

for i in range(0, len(data_map)):
    folium.Circle(
        location=[data_map.iloc[i]['latitude'], data_map.iloc[i]['longitude']] # latitude 위도,경도 
        ,color='crimson', 
        tooltip =   '<li><bold>Date : '+str(data_map.iloc[i]['date'])+
                    '<li><bold>Province : '+str(data_map.iloc[i]['province'])+
                    '<li><bold>City : '+str(data_map.iloc[i]['city'])+
                    '<li><bold>Contact_number : '+str(data_map.iloc[i]['contact_number'])+
                    '<li><bold>Num_Tracking : '+str(data_map.iloc[i]['Num_tracking']),
        radius=int(data_map.iloc[i]['Num_tracking'])).add_to(m)
```
